```python
import torch

x = torch.tensor(2.0, requires_grad=True)
y = x ** 2          # y = x²
z = y * 3           # z = 3x²

z.backward()        #Compute ∂z/∂x

print("x.grad =", x.grad)    #6x = 6*2 =12
```

z = 3y

∂z/∂y = 3

y = x²

∂y/∂x = 2x

now,  ∂z/∂x

 = ∂z/∂y × ∂y/∂x
        = 3 × (2x)
        = 6x



```python
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
y = x * 2       #y = [2, 4, 6]
loss = y.sum()   # scalar and loss = 2 + 4 + 6 = 12


loss.backward() #∂loss/∂x..“How does changing each element of x change loss?”


print(x.grad)
```

**Gradient accumulation**

```python
x = torch.tensor(2.0, requires_grad=True)

y1 = x ** 2
y1.backward()
print(x.grad)  # 4

y2 = x ** 3
y2.backward()
print(x.grad)  # 4 + 12 = 16 ❌
```

```python
x = torch.tensor(2.0, requires_grad=True)

y1 = x ** 2
y1.backward()
print(x.grad)  # 4

y2 = x ** 3

x.grad = None  # or x.grad.zero_()
y2.backward()
print(x.grad)  # 12 ✅
```

```python
x = torch.tensor(2.0, requires_grad=True)

y = x * 3
z = y.detach()
loss = z ** 2

loss.backward()
print(x.grad)
```
